```python
dataset_path = '../data/Annotations/'
```

```python
from utils import dataset
import matplotlib.pyplot as plt

train_set = dataset.ClothingDataset(dataset_path, train='train')
test_set = dataset.ClothingDataset(dataset_path, train='test')
```

# Train Set

```python
series = train_set.imgs['class'].value_counts(normalize=True).sort_index()
cnt = list(series*100)

explode = [0.2 if i > 5 else 0 for i in cnt]

labels = [f'Class: {label}' if expl > 0 else '' for label, expl in zip(list(series.index), explode)]

def percentage_value(value, threshold=5):
    if value > threshold:
        return f'{round(value, 2)}%'
    else:
        return ''

fig = plt.figure(figsize=(16, 8))
plt.pie(cnt, labels=labels, explode=explode, autopct=percentage_value)
fig.legend(list(series.index), ncol=3, fontsize='x-large', loc='center right')
fig.show()
```

# Test Set

```python
series = test_set.imgs['class'].value_counts(normalize=True).sort_index()
cnt = list(series*100)

explode = [0.2 if i > 5 else 0 for i in cnt]

labels = [f'Class: {label}' if expl > 0 else '' for label, expl in zip(list(series.index), explode)]

def percentage_value(value, threshold=5):
    if value > threshold:
        return f'{round(value, 2)}%'
    else:
        return ''

fig = plt.figure(figsize=(16, 8))
plt.pie(cnt, labels=labels, explode=explode, autopct=percentage_value)
fig.legend(list(series.index), ncol=3, fontsize='x-large', loc='center right')
fig.show()
```
